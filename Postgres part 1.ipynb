### **FUNCTION**

---

What is a function?

It is something that takes an input and generates an output eg, the concat function.

A function is a set of steps that creates a single value eg calculating sum.

---

#### **Scalar function**

* Scalar functions run against each individual row just like concat
* What they do, you give them input and they are going to run the function against each individual row.
* We get muiltiple outputs unlike in aggregate, where we get a single output.

---

#### **Aggregate Functions**

* Aggregate functions run against all the data to produce one output eg calculating the total sum of something
* The difference between them and scalar functions is that scalar run against individual rows.
* Agg operate on many records to produce one value, producing a summury while scalar on each recored independently.

---

* **COUNT()** -- counts rows in a specified table or view
* **MIN()** -- gets the minimum value in a set of values
* **MAX()** -- gets the maximum value in a set of values
* **SUM()** -- calculates the sum of values

#### **How Do we do that?**

1. **Filter out Nulls**

Because we cannot use comparison operators such as = or !=, to filter out null as we would get null as we saw in the above examples, we use the 'IS' operator.

##### **The *IS* keyword**

* The IS keyword allows us to filter on values that are null, not null, true or false.

* It's a special keyword that is reserverd when we want to check for Null and Not Null values using the where clouse.

Syntax:

> **SELECT * FROM \<table\> WHERE <field> = ' ' IS NOT FALSE or WHERE <field> IS TRUE**

* It is not a must for us to add this "IS NOT TRUE" at the end of or query but for explanation purposes that is why we added it there.

2. **Cleaning Up your Data**

This is when we want to replace the null values.

To do this, we would do something called null value substitution.

**NULL Value Substitution**

* This is the ability to replace null values to operate on data.

* We are basically making a null value not a null value, we are replacing it with something else.

##### **The COALESCE keyword**

* This is a keyword that allows us to do null value substitution

syntax:

> **SELECT COALESCE (\<column\>, 'empty') AS Colunm_alias FROM \<table\>....**

In the syntax above, what we are saying is that, in teh selected column - \<column\>, look for every null value and replace it with 'Empty' and because we are performing a function, we will need to rename the column as it will take the function's name if we dont.

* NOTE, Coalesce returns the first non-null value in a list.

* For instance, let's say we are performing null value substitution on multiple columns, it will go through each and every column and if it won't find any null values, it will return the original column. Its like the if else statement.

---

**Combined Coalsce**

This is when you have a table that has multiple columns that have null values, you can coalesce each column and then have a fallback.

syntax

> **SELECT COALESCE(\<column1>, \<column2\>, \<column3\>, 'Empty') AS combined_columns FROM \<table\>**

### **The LIKE keyword**

#### **Partial lookups**

This helps us especially in scenarios where we don't know what we are looking for or what we are searching for.

To achieve this, we use the like keyword

The syntax is:

> **SELECT first_name FROM employees WHERE first_name LIKE 'M%';**

This one, gets everyone who's name start with 'M', and the percentage sign means, anything else after that.

The LIKE operator in SQL is our key tool for navigating the textual part of data, allowing target searches within the data using wild cards to tune our searches.

For instance, a humanitarian aid worker could use it to quickly locate all NGOs with names that are related to water within a massive database.

There are two wildcards used in sql:

* **Underscore - "_"**
* * Represents a single character.
* * A search parttern like H_t will match with values like Hot, Hat and Hit, but would not match with  Heat because (_) specifies a single character

* **Percentage - "%"**
* * Rep multiple characters.
* * A search pattern like south% will match with values like south korea, south africa or southern since it represents any number of characters.

#### **Pattern Matching**

In order to use LIKE you need to build patterns to match.

Pattern matching happens with wild cards

**1. % at the end**

Matches any string starting with the given characters, for example, p% must start with p, be any length and can end with any character.

**2. % at the start**

Matches any string ending with the given characters, for example, %t can start with any character, can be any length, but must end with T, eg cat, pet, chart

**3. _ in place of one character**

Matches any single character in that position, for example, _at must contain only three characters and must end with 'at', eg Cat, pat, etc.

**4. % inside**

Matches any string that begins and ends with the given characters, and can be any length, for example c%t must start with c and contain ant number of characters, and must end with t, for example cat, cart, chart.

**World cards combined**

**_ at both ends**

Matches any string containing the given characters, three characters long and can start and end with any character, but must have A in the middle. eg Car, cat, pat, etc.

**% and _**

Using both % and _, we can limit strings further, for example, _a% matches with cat, cart.
_at can start with any single character that must be followed by a and can end with any number of characters.

```python
%%sql

/*
For cases where we are unsure whether what we are looking for is upper case of lower case, we use
the ILIKE keyword.
Let's say in our example we were unsure of the case of the name we are looking for, we would write them
in this format:

This will give us the results despite the case of the words we are looking for.
*/

SELECT *
FROM employees
WHERE first_name ILIKE 'g%ER'
LIMIT 20;
```
