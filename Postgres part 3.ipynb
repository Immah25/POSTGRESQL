### **JOINS**

A join combines columns from one table with those of another

Its a combination of one table from another table and combination happens on the link between two pieces of data, in our case, the empo_no in the employees table and the empo_no in the salaries table which happen to have a correlation

Joins take a column from one table that can map to the column of another table.

A join agg data from two tables but it does it based on taking one column from one table that can map to the column of anther table in our case, emp_no to emp_no.

The most common approach is to link primary key to foreign key

Joins usually happen from a primary key and foreign key but there instances it happens btween two columns of different tables which have the same data, for instance cities, where the column cities exists in both tables.

In such cases we use the colunms to create the link

It is not a primary to foreign key, its just two tables that happen to have two tables with the same data.

But in most cases, probably 99% cases, we will be doing a primary to foreign key links.

The above join we have done is just a basic join, let's look at the different ways we can perform join operations.

### 1. **Inner Join**

An example of inner join is what we did in the above example.

If we have a column that links the column from table A to table B, we can use an inner join.

The inner join finds the intersection of two datsets based on a link and it returns us the subsets.

Let's break it down.

If table A has a 1000 records and table B has a 1000 records and we link emp_no from table A to table_B and its a primary key to a foreign key relationship, if table B has keys for 500 salaries of table A, then the match is going to be 500 matches.

So we are matching columns from table Ato table and the intersection that we will match from A to B is what will be shown.

In the previous examples, we did the join with the where clouse which is for filtering data but we have a better way of doing it, using INNER JOIN from SQL.

We have a syntax called **INNER JOIN** for this.

Let's break it down.

When we do a join, we got to decide what is going to be on the left side and what is going to be on the right side, meaning, where am i selecting from and what am i trying to join it with.

If the question is we wanna have the salaries from employees, then we are going to select from employees and we are going to add salaries to it because employee is the main part of the data that we are trying to get.

So deciding what is going to be to the left and right is very important, where am i selecting from and what am i joining to.

Let's do an example...

---

```python
%%sql

SELECT table_name, 
        string_agg(column_name, ' , ' ORDER BY ordinal_position)
FROM information_schema.columns
WHERE table_schema='public'
GROUP BY table_name
ORDER BY table_name;
```

```python
%%sql

-- We are going to select from employees table and then joing to salaries table to get employees information and there salaries.

SELECT a.emp_no,
        CONCAT(a.first_name, ' ', a.last_name) 
        AS "name",
        b.salarY
FROM employees as a 
INNER JOIN salaries as b ON a.emp_no=b.emp_no
ORDER BY a.emp_no ASC
LIMIT 30;
```

---

So, how do we read the above code?

When we do an inner join as we did above, we are simply saying, link the salaries table to the employees table and do it on the emp_no

The inner join syntax is generally considered a best practice, it's more readable than using the where syntax and it shows us what's being joined.

The results always come back unsorted so when you do a join, make sure you order by emp_no.

Let's do another example

Let's say we wanted to know the salary of an employee, but only the ones that led to a new title

So here we are basically linking three tables.

So we wanna know the raises from a promotion

We know that employees table is linked to salaries table through emp_no but title changes link to the salaries through the from date

**Key Insight:- Title change always follow 2 days after the raise** 

With this, we can be able to know how to right our query even though it's going to be abit more complex.



---

```python
%%sql

/*
Let's say we wanted to know the salary of an employee, but only the ones that led to a new title
*/

SELECT table_name,
        string_agg(column_name, ' , ' ORDER BY ordinal_position)
FROM information_schema.columns
WHERE table_schema='public'
GROUP BY table_name
ORDER BY table_name

```

```python
%%sql 
SELECT a.emp_no,
        CONCAT(a.first_name, ' ', a.last_name)
        AS "name",
        b.salary,
        c.title,
        c.from_date 
        AS "Promoted on"
FROM employees AS a
INNER JOIN salaries AS b ON a.emp_no = b.emp_no
INNER JOIN titles AS c 
        ON b.emp_no=c.emp_no
        AND c.from_date = (b.from_date + INTERVAL '2 days')
        --The two keys from a COMPOSITE KEY.
ORDER BY a.emp_no ASC, b.from_date ASC

LIMIT 50;
```

---

**The ON keyword works just like the where keyword**

We were joining three tables using an employee number, but we realized that employee number alone wasn’t unique across all tables. An employee could appear multiple times—for instance, due to different roles or time periods. To avoid duplicates, we combined both employee number and from date as a composite key. In short, using both ensures each record is uniquely identified and prevents accidental duplicates in the future.

Let's say we wanted to know the original salary and salary at promotion..

Let's do it practically...

---

```python
%%sql

SELECT table_name,
        string_agg(column_name, ' , ' ORDER BY ordinal_position)
FROM information_schema.columns
WHERE table_schema = 'public'
GROUP BY table_name
ORDER BY table_name
```

```python
%%sql

/*
We wanna get the salary of an employee, the hire salary and the one that led to a new title, 
and we wanna know the raise from promotion
*/

SELECT a.emp_no,
        CONCAT(a.first_name, ' ', a.last_name)
        AS "name",
        COALESCE(b.from_date :: text, '-'),

        --We cast the date to string so that we can perform string operations
        --Otherwise, we'll get an error
        
        b.salary,
        COALESCE(c.title, 'No title change')
FROM employees AS a
INNER JOIN salaries AS b ON a.emp_no = b.emp_no
INNER JOIN titles AS c
        ON b.emp_no = c.emp_no
        AND (c.from_date = (b.from_date + INTERVAL '2 days')
        OR c.from_date = b.from_date)
ORDER BY a.emp_no ASC, b.from_date ASC
LIMIT 20;

```

### **SELF JOINS**

This is joining a table to it's self.

This is possible when a table has a foreign key referencing its primary key

Case example, let's say we wanted to see the supervisors name?

But we didn't want to see it by id because let's say we have 500, 000 employees and we have multiple level of supervisors and we wanted to see the supervisor's name besides the person because we want to see who they report to.

Self join is the solution.

It works the same way that the original inner join worked only that this time we are selecting from the same table twice as we see below.

Another difference is that there is no special syntax for self join.

The syntax would be something like:

> **SELECT a.id, a.name AS "employee", b.name as "supervisor name" FROM employee AS a, employee AS b WHERE a.supervisorid = b.id**

We could also do this with the inner join

> **SELECT a.id, a.name AS "employee", b.name as "supervisor name" FROM employee AS a INNER JOIN  employee AS b WHERE a.supervisorid = b.id**

Both of them would basically give us the same results because self join is the same thing as the inner join

Inner joins are just joining data and trying to find that middle intersection but a self join is just joining a table to it's self because a primary key and foreign key are in the same exact table

A self join is not a different type of join necessarilly, but it is a possibility for us to join a dataset to the same dataset as long as we have a primary key and a foreign key in the same table that reference the same exact information, in our case below being staff id and manger id.

So, whenever we find our self in a situation where we need to join a table to itself, we can use the self join.

Let's do it practically....

---

```python
%%sql

--Let's create a table to practice Self join

CREATE TABLE staff_hierarchy (
    staff_id SERIAL PRIMARY KEY,
    staff_name VARCHAR(100) NOT NULL,
    role VARCHAR(100),
    manager_id INT,

    FOREIGN KEY (manager_id) REFERENCES staff_hierarchy(staff_id)
);

INSERT INTO staff_hierarchy (staff_name, role, manager_id) VALUES
('Alice', 'Director', NULL),
('Brian', 'Team Lead', 1),
('Cynthia', 'Engineer', 2),
('David', 'Engineer', 2),
('Eva', 'Intern', 3);
```

```python
%%sql

SELECT column_name
FROM information_schema.columns
WHERE table_schema = 'public'
AND table_name = 'staff_hierarchy';

-- Let's say we wanted the staff name and there role
-- Self join way

SELECT  
    a.staff_name,
    b.role
FROM staff_hierarchy as a, staff_hierarchy as b
WHERE a.staff_id = b.manager_id
```

```python
%%sql

-- Inner join way

SELECT 
    a.staff_name,
    b.role
FROM staff_hierarchy as a
INNER JOIN staff_hierarchy as b ON a.staff_id = b.manager_id;
```

### **Outer Join**

What if we need the rows that don't match?

An OUTER JOIN is used when you want to keep rows even if there is no match in the other table.

They add the data that don't have a match.

**Simple way to think about it:**

- INNER JOIN :- Only matching data
- OUTER JOIN :- matching + non-matching data

**Why it matters:**

Sometimes data is incomplete, for example:

- Some employees don't have managers
- Some customers haven't made orders
- Some products haven't been sold

If we use inner join, we loose these rows, if we use outer join, we keep them.



---

#### **Types of outer join**

1. **LEFT JOIN (most common)**

Keep everything from the left table even if not match

Records that don't match will appear as NULL.

Left joins add the data that don't have a match from the left table **from the table that we put in the from clouse.**

When we are talking about on the left side of the join, we are talking about the table that is in the from statement, when we are talking about the right side of the join, we are talking about the table that is beign joined.

**Syntax..**

> **SELECT * FROM \<table\> AS a LEFT [OUTER] JOIN \<table\> AS b ON a.id = b.id**

We can decide to type LEFT OUTER JOIN or LEFT JOIN, they mean the same thing.

In our case we'll work with LEFT JOIN.

What this left join is goin to do is that it's going to say, let me find everything where a.id=b.id and also everywhere i couldn't find and a.id, i am just going to return that data to you

Some values may contain null because we can't find the correlation with b.

**Any value that does not match is made to be null**

We can decide to coalesce these null values 

Let's say we wanted to know how many employees aren't managers?

Let's do it practically...

---


```python
%%sql

SELECT table_name,
        string_agg(column_name, ' , ' ORDER BY ordinal_position)
FROM information_schema.columns
WHERE table_schema='public'
GROUP BY table_name
ORDER BY table_name
```

```python
%%sql
--Option 1
--Let's say we wanted to know how many employees aren't managers?

SELECT COUNT(emp.emp_no) AS "No of people not managers"
        --CONCAT(emp.first_name, ' ', emp.last_name),
        --tit.title

FROM employees AS emp
LEFT JOIN titles AS tit 
        ON emp.emp_no = tit.emp_no
        AND tit.title = 'Manager'
WHERE tit.title IS NULL
LIMIT 20;
```

```python
%%sql
--Option 2
--Linking the employees table with the department manager table

SELECT COUNT(emp.emp_no) AS "No of emp not managers"
        --CONCAT(emp.first_name, ' ', emp.last_name),
        --COALESCE(dep.dept_no, 'Not a manger')
FROM employees AS emp
LEFT JOIN dept_manager as dep ON emp.emp_no = dep.emp_no
WHERE dep.emp_no IS NULL
LIMIT 50;
```

---

So the left join allows us to take into account everyone that didn't match against department manager, in our case.

We could have done thios with multiple steps and then do the math but this is a simple way of achieving it, SQL does the hard lifting for us.

```python
%%sql

SELECT table_name,
        string_agg(column_name, ' , ' ORDER BY ordinal_position)
FROM information_schema.columns
WHERE table_schema='public'
GROUP BY table_name
ORDER BY table_name
```

```python
%%sql

SELECT a.emp_no,
        CONCAT(a.first_name, ' ', a.last_name)
        AS "name",
        COALESCE(b.from_date :: text, '-')
        AS "from_date",

        --We cast the date to string so that we can perform string operations
        --Otherwise, we'll get an error
        
        b.salary,
        COALESCE(c.title, 'No title change') 
        AS "Title"
FROM employees AS a
INNER JOIN salaries AS b ON a.emp_no = b.emp_no
LEFT JOIN titles AS c
        ON b.emp_no = c.emp_no
        AND (c.from_date = (b.from_date + INTERVAL '2 days')
        OR c.from_date = b.from_date)
ORDER BY a.emp_no ASC, b.from_date ASC
LIMIT 50;
```

Its saying heey, anything that is null on the left side if it doesn't match this criteria just show it as well.

### **Right Outer Join**

This keeps everything from the right table

Its adds the data that don't have a match from the right table.

**Not used often, we usually swap tables instead.**

It is the opp of left join

Syntax:

> **SELECT * FROM \<table A\> AS a RIGHT [OUTER] JOIN \<table B\> AS b ON a.id = b.id**

The syntax is the same except it uses right outer join

Any value that does not match is made to be null.

---

The difference between the two is that the LEFT JOIN keeps every thing from the left table, the table in the FROM statement and the RIGHT JOIN keeps everything from the right table, the table that is being joined.

In most cases, RIGHT JOIN is not used, people just swap the tables.

### **Joins we won't see often** 

There are more join techniques but they aren't used often. 

---

1. **Cross Join**

The cross join creates a combination of every row with every other row

Its called a cartesian product

Its a combination of everything

Let's create a table that we can do practically....

```python
%%sql

CREATE TABLE "cartesianA"(id INT);
CREATE TABLE "cartesianB"(id INT);
```

```python
%%sql

INSERT INTO "cartesianA" VALUES(1);
INSERT INTO "cartesianA" VALUES(2);
INSERT INTO "cartesianA" VALUES(3);

INSERT INTO "cartesianB" VALUES(1);
INSERT INTO "cartesianB" VALUES(2);
```

```python
%%sql

SELECT * 
FROM "cartesianA"
CROSS JOIN "cartesianB";
```

---

As you see in the above example, we are cross joining every value from A to every value from B.

Its returns us the cartesian product. The combination of every row and every other row

---

2. **The full outer Join**

It returns results from both whether they match or not.

Let's see how that would work with our cartesian table....

```python
%%sql

INSERT INTO "cartesianA" VALUES (4);
INSERT INTO "cartesianA" VALUES (5);
INSERT INTO "cartesianB" VALUES (20);
INSERT INTO "cartesianB" VALUES (30);
```

```python
%%sql

SELECT *
FROM "cartesianA" AS a 
FULL JOIN "cartesianB" AS b ON a.id = b.id
```

---

From the above, we see that everything that is not matched is returned

This is less common. Most cases involve left or right join.
