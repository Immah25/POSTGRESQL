### **Interval**

Interval allowas us to write queries in a way that mirrors language

It can store and manipulate a period of time in years, months, days, hours, minutes, seconds, etc.

For instance, let's say we wanted the date of purchase for thelast 30 days 

We could use the following query

> **SELECT * FROM orders WHERE purchaseDate <= now() - interval '30 days'**

Minus 30 days before means that if the purchase date is smaller than or equals to any purchase that happened 30 days prior in that 30 days period return them to us.

In short it returns purchase orders 30 days before the given date.

**The identifiers we get are, Years, Months, Days, Hours and Minutes.**

We can say 30 days, 30 years, 30 hours, 30 minutes etc

Its is an easy way to subtract and add to dates to know how far back or how far ahead you should be looking

We can say:

* **INTERVAL '1 year 2 months 3 days';**
* **INTERVAL '2 weeks ago';**
* **INTERVAL '1 year 3 hours 20 minutes'**

**And because we are using the interval keyword when we tell potgres to subtract or add to a date or timestamp, its going to do the calculation of all that math for us**

So we don't have to go ahead and calculate eg purchases for two weeks ago etc and do all that manipulation on a timestamp. The interval keyword does all that for us.

We can extract within an interval also:

> **SELECT EXTRACT(YEAR FROM INTERVAL '5 years 20 months')**

We will get an output of 6 years as 20 months is more that an year.

**When you want to do manipulation with dates towards the future or the past, always remember you should use intervals**
